```python
# Standard imports
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python

train_titanic = pd.read_csv("train.csv")
train_titanic.head()
```

```python
X = train_titanic.drop("Survived", axis=1)
y = train_titanic["Survived"]
```

```python
# Let's remove all unnecessary columns
X = X.drop(["PassengerId", "Name", "Ticket", "Cabin", "Fare"], axis=1)
X.head()
```

```python
X.value_counts()
```

```python
X.isna().value_counts()
```

```python
# Replacing male and female values to 1 and 0
X["Sex"] = X["Sex"].replace(["male"], 1)
X["Sex"].head()
```

```python
X["Sex"] = X["Sex"].replace(["female"], 0)
X["Sex"].head()
```

```python
X.fillna("0")
```

```python
X["Embarked"].value_counts()
```

```python
# Replacing an embarked port to 0, 1, 2 values 
X["Embarked"] = X["Embarked"].replace(["S", "C", "Q"], [0, 1, 2])
X["Embarked"]
```

```python
# Filling NaN values with 0
X.fillna("0", inplace=True)
X
```

```python
X.isna().value_counts()
```

```python
# Let's revise the input data
X.head()
```

```python
y.head()
```

```python
# Let's split the data into train and test for fitting the model and checking the score
from sklearn.model_selection import train_test_split

np.random.seed(42)
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.2)
X_train.shape, X_test.shape, y_train.shape, y_test.shape

# Will be using RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
np.random.seed(42)

for i in range(10, 100, 10):
    print(f"Trying model with {i} estimators...")
    model = RandomForestClassifier(n_estimators=i).fit(X_train, y_train)
    print(f"Model accuracy on test set: {model.score(X_test, y_test) * 100}%")
    print("")
```

```python
from sklearn.model_selection import cross_val_score

cross_val_score(model, X, y)
```

```python
grid = {"n_estimators": [10, 100, 200, 500, 1000, 1200],
        "max_depth": [None, 5, 10, 20, 30],
        "max_features": ["auto", "sqrt"],
        "min_samples_split": [2, 4, 6],
        "min_samples_leaf": [1, 2, 4]}
```

```python
from sklearn.model_selection import RandomizedSearchCV

clf = RandomForestClassifier()

# Setup RandomizedSearchCV
rs_clf = RandomizedSearchCV(estimator=clf,
                            param_distributions=grid,
                            n_iter=20, # try 20 models total
                            cv=5, # 5-fold cross-validation
                            verbose=2) # print out results

# Fit the RandomizedSearchCV version of clf
rs_clf.fit(X_train, y_train);
```

```python
rs_clf.best_params_
```

```python
# Another hyperparameter grid similar to rs_clf.best_params_
grid_2 = {'n_estimators': [1200, 1500, 2000],
          'max_depth': [None, 5, 10],
          'max_features': ['auto', 'sqrt'],
          'min_samples_split': [4, 6],
          'min_samples_leaf': [1, 2]}
```

```python
from sklearn.model_selection import GridSearchCV

np.random.seed(42)

# Set n_jobs to -1 to use all cores (NOTE: n_jobs=-1 is broken as of 8 Dec 2019, using n_jobs=1 works)
clf2 = RandomForestClassifier()

# Setup GridSearchCV
gs_clf2 = GridSearchCV(estimator=clf,
                      param_grid=grid_2,
                      cv=5, # 5-fold cross-validation
                      verbose=2) # print out progress

# Fit the RandomizedSearchCV version of clf
gs_clf2.fit(X_train, y_train);
```

```python
gs_clf2.best_params_
```

```python
# Training model with best params
from sklearn.ensemble import RandomForestClassifier

best_model = RandomForestClassifier(max_depth=5,
                               max_features="sqrt",
                               min_samples_leaf=2,
                               min_samples_split=6,
                               n_estimators=1500)
best_model.fit(X, y)
```
